```python
#importing all important libraries
import pandas as pd # pandas library to work with dataframes
import numpy as np # numpy library to work with numerical or array functions
import matplotlib.pyplot as plt # matplotib is visualization library
import seaborn as sns # seaborn visualization library
%matplotlib inline 
# load data
file_path = r"D:\WebNeuralInfotech\Webscrapping\BeautifulSoup\books_data.csv"
books_df = pd.read_csv(file_path)
```

```python
#emoving duplicates
new_books_df=books_df.drop_duplicates()
new_books_df
```

```python
#Ensure the Column Contains Only Strings
new_books_df['Price'] = new_books_df['Price'].astype(str)
new_books_df
```

```python
#remove the unwanted characters using .str.replace():
new_books_df['Price'] = new_books_df['Price'].str.replace('Â£', '').replace('', float('NaN')).astype(float)
new_books_df
```

```python
# Check the data type
print(new_books_df['Availability'].dtype)
```

```python
# Strip leading/trailing whitespace and ensure the column is a string
new_books_df['Availability'] = new_books_df['Availability'].fillna('').astype(str).str.strip()
# Use .str.contains with case-insensitivity
new_books_df['Availability'] = new_books_df['Availability'].str.contains('In stock', case=False)
new_books_df
```

```python
new_books_df['Rating'] = new_books_df['Rating'].apply(lambda x: {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}.get(x, 0))
new_books_df
```

```python
rating_counts = new_books_df['Rating'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
rating_counts.plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Distribution of Book Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Books')
plt.xticks(rotation=0)
```

```python
#To see Distribution of Prices we use Histogram
plt.figure(figsize=(10, 6))
plt.hist(new_books_df['Price'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Book Prices')
plt.xlabel('Price (in £)')
plt.ylabel('Frequency')
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=new_books_df['Price'], color='skyblue')
plt.title('Price Distribution of Books')
plt.xlabel('Price (in £)')
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=new_books_df['Rating'], color='skyblue')
plt.title('Rating of Books')
plt.xlabel('rating')
```
